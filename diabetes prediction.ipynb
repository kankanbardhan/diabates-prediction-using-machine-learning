```python
X=df1.drop("Outcome",axis=1)
Y=df["Outcome"]
```

```python
X
```

```python
Y
```

```python
X.replace(0,df1.mean(),inplace=True)
```

```python
X.head(20)
```

```python
X.describe()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
```

```python
X_train_std
```

```python
Y_train
```

```python
print(X_train_std.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
Y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression()
```

```python
Lr.fit(X_train_std,Y_train)
```

```python
Y_pred=Lr.predict(X_test_std)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Testing Accuracy =",accuracy_score(Y_pred,Y_test)*100)
print("Training Accuracy =",Lr.score(X_train_std,Y_train)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train_std,Y_train)
```

```python
Y_Pred=dtc.predict(X_test_std)
```

```python
print("Testing Accuracy =",accuracy_score(Y_Pred,Y_test)*100)
print("Training Accuracy =",dtc.score(X_train_std,Y_train)*100)
```

##### In the above DecisionTreeClassifier algorithm, Overfitting Happens. Because the difference between Training and testing accuracy is very  much high as compared to other Algorithms.

```python
from sklearn.naive_bayes import GaussianNB
GB=GaussianNB()
```

```python
GB.fit(X_train_std,Y_train)
```

```python
Y_Predict=GB.predict(X_test_std)
```

```python
print("Testing Accuracy =",accuracy_score(Y_Predict,Y_test)*100)
print("Training Accuracy =",GB.score(X_train_std,Y_train)*100)
```

```python
#Import svm model
from sklearn.svm import SVC  
clf = SVC()
clf.fit(X_train_std, Y_train)
y_pred = clf.predict(X_test_std)
print("Testing Accuracy =",accuracy_score(y_pred,Y_test)*100)
print("Training Accuracy =",clf.score(X_train_std,Y_train)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train_std, Y_train)
y_predict = knn.predict(X_test_std)
print("Testing Accuracy =",accuracy_score(y_predict,Y_test)*100)
print("Training Accuracy =",knn.score(X_train_std,Y_train)*100)
```

# Apply Deep Learning

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(hidden_layer_sizes=(8,8,8,8)) # 8 Neuron in each hidden layer.we can change the hidden
mlp.fit(X_train_std, Y_train)                   # layers number and the size of the Neuron.
y_predi = mlp.predict(X_test_std)
print("Testing Accuracy =",accuracy_score(y_predi,Y_test)*100)
print("Training Accuracy =",mlp.score(X_train_std,Y_train)*100)
```
